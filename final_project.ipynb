Define Functions for Getting Data From and Setting Data Into Databases

```python
def get_sql_dataframe(sql_query, **args):
    '''Create a connection to the MySQL database'''
    conn_str = f"mysql+pymysql://{args['uid']}:{args['pwd']}@{args['hostname']}/{args['dbname']}"
    sqlEngine = create_engine(conn_str, pool_recycle=3600)
    connection = sqlEngine.connect()
    
    '''Invoke the pd.read_sql() function to query the database, and fill a Pandas DataFrame.'''
    dframe = pd.read_sql(text(sql_query), connection);
    connection.close()
    
    return dframe
    

def set_dataframe(df, table_name, pk_column, db_operation, **args):
    '''Create a connection to the MySQL database'''
    conn_str = f"mysql+pymysql://{args['uid']}:{args['pwd']}@{args['hostname']}/{args['dbname']}"
    sqlEngine = create_engine(conn_str, pool_recycle=3600)
    connection = sqlEngine.connect()
    
    '''Invoke the Pandas DataFrame .to_sql( ) function to either create, or append to, a table'''
    if db_operation == "insert":
        df.to_sql(table_name, con=connection, index=False, if_exists='replace')
        connection.execute(text(f"ALTER TABLE {table_name} ADD PRIMARY KEY ({pk_column});"))
            
    elif db_operation == "update":
        df.to_sql(table_name, con=connection, index=False, if_exists='append')
    
    connection.close()


def get_mongo_client(**args):
    '''Validate proper input'''
    if args["cluster_location"] not in ['atlas', 'local']:
        raise Exception("You must specify either 'atlas' or 'local' for the cluster_location parameter.")
    
    else:
        if args["cluster_location"] == "atlas":
            connect_str = f"mongodb+srv://{args['user_name']}:{args['password']}@"
            connect_str += f"{args['cluster_name']}.{args['cluster_subnet']}.mongodb.net"
            client = pymongo.MongoClient(connect_str, tlsCAFile=certifi.where())
            
        elif args["cluster_location"] == "local":
            client = pymongo.MongoClient("mongodb://localhost:27017/")
        
    return client


def get_mongo_dataframe(mongo_client, db_name, collection, query):
    '''Query MongoDB, and fill a python list with documents to create a DataFrame'''
    db = mongo_client[db_name]
    dframe = pd.DataFrame(list(db[collection].find(query)))
    dframe.drop(['_id'], axis=1, inplace=True)
    mongo_client.close()
    
    return dframe


def set_mongo_collections(mongo_client, db_name, data_directory, json_files):
    db = mongo_client[db_name]
    
    for file in json_files:
        db.drop_collection(file)
        json_file = os.path.join(data_directory, json_files[file])
        with open(json_file, 'r') as openfile:
            json_object = json.load(openfile)
            file = db[file]
            result = file.insert_many(json_object)
        
    mongo_client.close()
```

### Query Top 5 User–Movie Combinations by Total Comments


```python
# QUERY 1: Total Comments per User per Movie 
# This query joins fact_comments, dim_users, and dim_movies to find how many comments each user has made per movie. It shows the top 5 user–movie combinations with the most comments.

sql_comments_per_user_movie = """
SELECT
    u.user_key,
    u.name AS user_name,
    m.movie_key,
    m.title AS movie_title,
    COUNT(c.comment_key) AS total_comments
FROM
    fact_comments c
JOIN
    dim_users u ON c.user_key = u.user_key
JOIN
    dim_movies m ON c.movie_key = m.movie_key
GROUP BY
    u.user_key, u.name, m.movie_key, m.title
ORDER BY
    total_comments DESC
LIMIT 20;
"""

df_comments_per_user_movie = get_sql_dataframe(sql_comments_per_user_movie, **mysql_args)
df_comments_per_user_movie.head(5)
```

### Compute Average IMDb Rating per User Based on Their Comments


```python
#QUERY 2: Average IMDb Rating per User
# This query computes the average IMDb rating of all movies commented on by each user. It joins fact_comments, dim_users, and dim_imdb_top_1000 and returns the top 5 users ranked by their average movie rating.

sql_avg_rating_per_user = """
SELECT
    u.user_key,
    u.name AS user_name,
    COUNT(c.comment_key) AS total_comments,
    AVG(m.imdb_rating) AS avg_imdb_rating
FROM
    fact_comments c
JOIN
    dim_users u ON c.user_key = u.user_key
JOIN
    dim_imdb_top_1000 m ON c.movie_key = m.movie_key
GROUP BY
    u.user_key, u.name
ORDER BY
    avg_imdb_rating DESC
LIMIT 20;
"""

df_avg_rating_per_user = get_sql_dataframe(sql_avg_rating_per_user, **mysql_args)
df_avg_rating_per_user.head(5)
```

### Analyze Sakila Customers by Total Comments and Average IMDb Ratings


```python
#QUERY 3: Comments and Average Ratings by Sakila Customers
# This query integrates the Sakila database by linking users to Sakila customers (matching by user_key). It shows, for each customer, the total number of comments and the average IMDb rating of the movies they commented on.

sql_comments_sakila = """
SELECT
    cust.customer_id,
    cust.first_name,
    cust.last_name,
    COUNT(c.comment_key) AS total_comments,
    AVG(m.imdb_rating) AS avg_movie_rating
FROM
    fact_comments c
JOIN
    dim_users u ON c.user_key = u.user_key
JOIN
    dim_imdb_top_1000 m ON c.movie_key = m.movie_key
JOIN
    sakila.customer cust ON cust.customer_id = u.user_key
GROUP BY
    cust.customer_id, cust.first_name, cust.last_name
ORDER BY
    total_comments DESC
LIMIT 20;
"""

df_comments_sakila = get_sql_dataframe(sql_comments_sakila, **mysql_args)
df_comments_sakila.head(5)
```

### Initialize PySpark and Create Database


```python
import os
import pandas as pd
import numpy as np
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, count, avg, explode, split


spark = SparkSession.builder \
    .appName("MovieDLH") \
    .config("spark.sql.warehouse.dir", os.path.abspath("spark-warehouse")) \
    .getOrCreate()
spark.sparkContext.setLogLevel("OFF")

dest_database = "movie_dlh"
spark.sql(f"DROP DATABASE IF EXISTS {dest_database} CASCADE")
spark.sql(f"CREATE DATABASE IF NOT EXISTS {dest_database}")
```

### Convert Pandas DataFrames to Spark DataFrames


```python
spark_dim_movies      = spark.createDataFrame(df_dim_movies)
spark_dim_imdb        = spark.createDataFrame(df_imdb)
spark_dim_users       = spark.createDataFrame(df_dim_users)
spark_fact_comments   = spark.createDataFrame(df_fact_comments)
spark_dim_actor       = spark.createDataFrame(df_dim_actor)
spark_dim_film        = spark.createDataFrame(df_dim_film)
spark_fact_film_actor = spark.createDataFrame(df_fact_film_actor)
spark_dim_date        = spark.createDataFrame(df_dim_date)
```

### Save all Bronze-layer tables as managed Spark tables


```python
spark_dim_movies.write.saveAsTable(f"{dest_database}.bronze_dim_movies", mode="overwrite")
spark_dim_imdb.write.saveAsTable(f"{dest_database}.bronze_dim_imdb", mode="overwrite")
spark_dim_users.write.saveAsTable(f"{dest_database}.bronze_dim_users", mode="overwrite")
spark_fact_comments.write.saveAsTable(f"{dest_database}.bronze_fact_comments", mode="overwrite")
spark_dim_actor.write.saveAsTable(f"{dest_database}.bronze_dim_actor", mode="overwrite")
spark_dim_film.write.saveAsTable(f"{dest_database}.bronze_dim_film", mode="overwrite")
spark_fact_film_actor.write.saveAsTable(f"{dest_database}.bronze_fact_film_actor", mode="overwrite")
spark_dim_date.write.saveAsTable(f"{dest_database}.bronze_dim_date", mode="overwrite")
```

### Create Silver-layer table by enriching fact tables with dimension data and basic aggregates


```python
#Silver Layer: Comments with Users & Movies
spark_silver_comments = spark_fact_comments \
    .join(spark_dim_movies, "movie_key") \
    .join(spark_dim_users, "user_key")
spark_silver_comments.write.saveAsTable(f"{dest_database}.silver_comments", mode="overwrite")

#Silver Layer: Movies enriched with IMDb ratings
spark_silver_movies = spark_dim_movies.join(
    spark_dim_imdb[['movie_key', 'imdb_rating', 'metascore', 'imdb_votes']],
    on='movie_key',
    how='left'
)
spark_silver_movies.write.saveAsTable(f"{dest_database}.silver_movies", mode="overwrite")

#Silver Layer: Users with aggregated info (comments count)
spark_silver_users = spark_fact_comments.groupBy("user_key") \
    .agg(count("comment_key").alias("num_comments"))
spark_silver_users = spark_silver_users.join(spark_dim_users, "user_key")
spark_silver_users.write.saveAsTable(f"{dest_database}.silver_users", mode="overwrite")

#Silver Layer: Film-Actor enriched
spark_silver_film_actor = spark_fact_film_actor \
    .join(spark_dim_actor, "actor_key") \
    .join(spark_dim_film, "film_key")
spark_silver_film_actor.write.saveAsTable(f"{dest_database}.silver_film_actor", mode="overwrite")
```

### Gold Layer: aggregate silver tables for reporting


```python
#Gold: Movie Comments Aggregation
gold_movie_comments = spark_silver_comments.groupBy("movie_key", "title") \
    .agg(count("comment_key").alias("num_comments")) \
    .orderBy(col("num_comments").desc())
gold_movie_comments.write.saveAsTable(f"{dest_database}.gold_movie_comments", mode="overwrite")
gold_movie_comments.write.mode("overwrite").parquet(f"{dest_database}/gold/movie_comments")

#Gold: User Average Ratings & Comment Counts
gold_user_avg_rating = spark_silver_comments.join(
    spark_dim_imdb.select("movie_key", "imdb_rating"),
    on="movie_key",
    how="left"
).groupBy("user_key", "name") \
 .agg(
    avg("imdb_rating").alias("avg_imdb_rating"),
    count("comment_key").alias("num_comments")
 ).orderBy(col("avg_imdb_rating").desc())
gold_user_avg_rating.write.saveAsTable(f"{dest_database}.gold_user_avg_rating", mode="overwrite")
gold_user_avg_rating.write.mode("overwrite").parquet(f"{dest_database}/gold/user_avg_rating")

#Gold: Genre-Level Aggregation
gold_genre_summary = spark_silver_movies.withColumn(
    "genre", explode(split(col("genres"), ", "))
).groupBy("genre") \
 .agg(
    avg("imdb_rating").alias("avg_rating"),
    count("movie_key").alias("num_movies")
 ).orderBy(col("num_movies").desc())
gold_genre_summary.write.saveAsTable(f"{dest_database}.gold_genre_summary", mode="overwrite")
gold_genre_summary.write.mode("overwrite").parquet(f"{dest_database}/gold/genre_summary")

```
